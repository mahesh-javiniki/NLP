# Parts of Speech (POS) Tagging in NLP

## Introduction to POS Tagging

**Parts of Speech (POS) Tagging** is the process of assigning grammatical categories (tags) to each word in a text. It's a fundamental task in Natural Language Processing that helps computers understand the syntactic structure and meaning of sentences.

### What is POS Tagging?

POS tagging involves labeling each word in a sentence with its appropriate part of speech based on both its definition and its context. For example:
- **Noun (NN)**: Person, place, thing, or idea
- **Verb (VB)**: Action or state of being
- **Adjective (JJ)**: Describes a noun
- **Adverb (RB)**: Describes a verb, adjective, or another adverb
- **Pronoun (PRP)**: Replaces a noun
- **Preposition (IN)**: Shows relationship between words
- **Conjunction (CC)**: Connects words or phrases
- **Interjection (UH)**: Expresses emotion

### Why is POS Tagging Important?

1. **Text Analysis**: Understanding sentence structure and grammar
2. **Information Extraction**: Identifying entities and relationships
3. **Machine Translation**: Improving translation accuracy
4. **Speech Recognition**: Disambiguating words with multiple meanings
5. **Question Answering**: Understanding query intent
6. **Sentiment Analysis**: Identifying opinion-bearing words
7. **Named Entity Recognition**: Identifying proper nouns

### POS Tag Set

NLTK uses the Penn Treebank POS tag set, which includes 36 different tags. Some common ones:
- **NN**: Noun, singular
- **NNS**: Noun, plural
- **NNP**: Proper noun, singular
- **NNPS**: Proper noun, plural
- **VB**: Verb, base form
- **VBD**: Verb, past tense
- **VBG**: Verb, gerund/present participle
- **VBN**: Verb, past participle
- **VBP**: Verb, non-3rd person singular present
- **VBZ**: Verb, 3rd person singular present
- **JJ**: Adjective
- **JJR**: Adjective, comparative
- **JJS**: Adjective, superlative
- **RB**: Adverb
- **RBR**: Adverb, comparative
- **RBS**: Adverb, superlative

## Experiment 1: Basic POS Tagging

Let's start with a simple example to understand how POS tagging works.

```python
# Simple sentence for POS tagging
text = "The quick brown fox jumps over the lazy dog"

# Tokenize the text
tokens = word_tokenize(text)

# Perform POS tagging
pos_tags = pos_tag(tokens)

print("Original Text:", text)
print("\nTokens:", tokens)
print("\nPOS Tags:")
print(pos_tags)

# Display in a more readable format
print("\n" + "="*50)
print("Word-by-Word POS Tagging:")
print("="*50)
for word, tag in pos_tags:
    print(f"{word:15} --> {tag}")
```

### ðŸ“Š Observations - Experiment 1:

1. **Determiners (DT)**: "The" appears twice and is correctly tagged as a determiner
2. **Adjectives (JJ)**: "quick", "brown", and "lazy" are tagged as adjectives describing nouns
3. **Nouns (NN)**: "fox" and "dog" are singular nouns
4. **Verbs (VBZ)**: "jumps" is tagged as a 3rd person singular present verb
5. **Preposition (IN)**: "over" is correctly identified as a preposition
6. **Key Learning**: POS tagging considers both the word itself and its context in the sentence
7. **Accuracy**: NLTK's default tagger works well with simple, grammatically correct sentences

## Experiment 2: Understanding POS Tag Meanings

Let's explore what each POS tag means and get detailed information about tags.

```python
# Function to explain POS tags
def explain_tag(tag):
    """Get detailed information about a POS tag"""
    try:
        nltk.help.upenn_tagset(tag)
    except:
        print(f"No information available for tag: {tag}")

# Let's understand some common tags
print("Understanding Common POS Tags:")
print("="*60)

# Explain specific tags
tags_to_explain = ['NN', 'VBZ', 'JJ', 'DT', 'IN', 'RB', 'PRP']

for tag in tags_to_explain:
    print(f"\n{'='*60}")
    print(f"Tag: {tag}")
    print("="*60)
    explain_tag(tag)
```

### ðŸ“Š Observations - Experiment 2:

1. **Tag Hierarchy**: POS tags follow a systematic naming convention (e.g., VB, VBD, VBG, VBN, VBP, VBZ all relate to verbs)
2. **Granularity**: The Penn Treebank tagset provides fine-grained distinctions (e.g., different verb forms)
3. **Context Dependency**: Understanding tags helps interpret why certain words get specific tags
4. **Documentation**: NLTK provides built-in documentation for each tag using `upenn_tagset()`
5. **Practical Use**: Knowing tag meanings is crucial for filtering words by type (e.g., extracting only nouns)

## Experiment 3: POS Tagging on Complex Sentences

Let's test POS tagging on more complex sentences with different grammatical structures.

```python
# Different types of sentences
sentences = [
    "I am learning Natural Language Processing at university.",
    "She has been working on machine learning projects since 2020.",
    "The conference will be held in New York next month.",
    "Wow! That was an absolutely amazing performance!"
]

for i, sentence in enumerate(sentences, 1):
    print(f"\n{'='*70}")
    print(f"Sentence {i}: {sentence}")
    print('='*70)
    
    # Tokenize and tag
    tokens = word_tokenize(sentence)
    tags = pos_tag(tokens)
    
    # Display results
    for word, tag in tags:
        print(f"{word:20} --> {tag}")
    
    # Count different POS types
    pos_counts = {}
    for word, tag in tags:
        pos_counts[tag] = pos_counts.get(tag, 0) + 1
    
    print(f"\n{'POS Summary:'}")
    for tag, count in sorted(pos_counts.items()):
        print(f"  {tag}: {count}")
```

### ðŸ“Š Observations - Experiment 3:

1. **Compound Tenses**: Words like "has been working" are tagged separately (VBZ-VBN-VBG), showing how tagger handles complex verb phrases
2. **Proper Nouns (NNP)**: "Natural Language Processing", "New York" are correctly identified as proper nouns
3. **Gerunds (VBG)**: "learning", "working" are tagged as VBG (present participle/gerund)
4. **Auxiliary Verbs**: "am", "has", "will" are correctly distinguished from main verbs
5. **Interjections (UH)**: "Wow" is tagged as an interjection, showing emotion detection
6. **Adverbs (RB)**: "absolutely" modifying "amazing" is correctly tagged
7. **Time References**: "since 2020", "next month" show how temporal expressions are tagged
8. **Punctuation**: Each punctuation mark gets its own tag

## Experiment 4: Ambiguity in POS Tagging

Some words can function as different parts of speech depending on context. Let's explore this.

```python
# Words that can have different POS tags based on context
ambiguous_examples = [
    ("I will book a flight tomorrow.", "book as verb"),
    ("I am reading a book.", "book as noun"),
    ("The light was bright.", "light as noun"),
    ("Please light the candle.", "light as verb"),
    ("They run every morning.", "run as verb"),
    ("We went for a run.", "run as noun"),
    ("She can dance beautifully.", "can as modal verb"),
    ("Open the can of beans.", "can as noun"),
    ("The dog is fast asleep.", "fast as adverb"),
    ("He is a fast runner.", "fast as adjective")
]

print("AMBIGUITY IN POS TAGGING")
print("="*70)

for sentence, description in ambiguous_examples:
    tokens = word_tokenize(sentence)
    tags = pos_tag(tokens)
    
    print(f"\n{description.upper()}")
    print(f"Sentence: {sentence}")
    print("Tags:", end=" ")
    for word, tag in tags:
        print(f"{word}/{tag}", end=" ")
    print()
```

### ðŸ“Š Observations - Experiment 4:

1. **Context Matters**: The same word gets different tags based on its role in the sentence
2. **"book"**: Tagged as VB (verb) when action, NN (noun) when object
3. **"light"**: Can be noun (NN) or verb (VB) depending on usage
4. **"run"**: Verb form (VBP) vs noun form (NN)
5. **"can"**: Modal verb (MD) for ability vs noun (NN) for container
6. **"fast"**: Adverb (RB) when modifying verb, adjective (JJ) when modifying noun
7. **Disambiguation**: POS tagging helps resolve these ambiguities using context
8. **Accuracy**: NLTK's tagger correctly identifies most context-dependent meanings
9. **Challenge**: Some ambiguous cases might require more sophisticated models

## Experiment 5: Extracting Specific POS Categories

Learn how to filter and extract specific types of words based on their POS tags.

### ðŸ“Š Observations - Experiment 5:

1. **Prefix Matching**: Using `tag.startswith()` captures all variations (NN, NNS, NNP, NNPS)
2. **Nouns Dominance**: Nouns typically make up a large portion of content words in technical text
3. **Adjective-Noun Pairs**: Adjectives like "sophisticated", "complex" provide descriptive context
4. **Verb Forms**: Multiple verb forms appear (are, is, working, develop, transform)
5. **Practical Application**: This technique is useful for:
   - Keyword extraction (nouns)
   - Sentiment analysis (adjectives, adverbs)
   - Action identification (verbs)
   - Entity extraction (proper nouns)
6. **Text Characteristics**: Technical texts have higher noun density

## Experiment 6: POS Tagging with Different Taggers

NLTK provides multiple POS taggers. Let's explore and compare them.

```python
from nltk.tag import DefaultTagger, UnigramTagger, BigramTagger, TrigramTagger
from nltk.corpus import brown

# Download brown corpus for training
nltk.download('brown')
nltk.download('universal_tagset')

# Test sentence
test_sentence = "The cat is sleeping on the comfortable couch"
test_tokens = word_tokenize(test_sentence)

print("COMPARING DIFFERENT POS TAGGERS")
print("="*70)
print(f"Test Sentence: {test_sentence}\n")

# 1. Default Perceptron Tagger (NLTK's default)
print("\n1. DEFAULT PERCEPTRON TAGGER:")
print("-"*70)
default_tags = pos_tag(test_tokens)
for word, tag in default_tags:
    print(f"{word:20} --> {tag}")

# 2. Universal Tagset (Simplified tagset)
print("\n2. UNIVERSAL TAGSET (Simplified):")
print("-"*70)
universal_tags = pos_tag(test_tokens, tagset='universal')
for word, tag in universal_tags:
    print(f"{word:20} --> {tag}")

# 3. Default Tagger (baseline - assigns most common tag)
print("\n3. DEFAULT TAGGER (Baseline - always tags as NN):")
print("-"*70)
default_tagger = DefaultTagger('NN')
baseline_tags = default_tagger.tag(test_tokens)
for word, tag in baseline_tags:
    print(f"{word:20} --> {tag}")

print("\n" + "="*70)
print("TAGGER COMPARISON SUMMARY:")
print("="*70)
print("Default Perceptron: Pre-trained, high accuracy, context-aware")
print("Universal Tagset: Simplified (12 tags), easier to work with")
print("Default Tagger: Baseline, no learning, assigns same tag to all")
```

### ðŸ“Š Observations - Experiment 6:

1. **Perceptron Tagger**: 
   - Most accurate, uses machine learning
   - Penn Treebank tagset (36 tags)
   - Pre-trained on large corpus
   
2. **Universal Tagset**:
   - Only 12 tags: NOUN, VERB, ADJ, ADV, PRON, DET, ADP, NUM, CONJ, PRT, '.', X
   - Easier for beginners and cross-language work
   - Less granular but more portable
   
3. **Default Tagger**:
   - Baseline for comparison
   - No intelligence, assigns same tag to everything
   - Useful as fallback in training custom taggers
   
4. **Trade-offs**: More tags = more information but more complexity
5. **Use Cases**: 
   - Universal for multilingual projects
   - Penn Treebank for detailed English analysis
   - Default as fallback in sequential taggers

## Experiment 7: Training a Custom POS Tagger

Learn how to train your own POS tagger using the Brown corpus.

```python
# Get tagged sentences from Brown corpus
tagged_sentences = brown.tagged_sents(tagset='universal')

# Split into training and testing data
train_size = int(len(tagged_sentences) * 0.8)
train_data = tagged_sentences[:train_size]
test_data = tagged_sentences[train_size:]

print(f"Total sentences: {len(tagged_sentences)}")
print(f"Training sentences: {len(train_data)}")
print(f"Testing sentences: {len(test_data)}")

# Example of tagged data
print("\n" + "="*70)
print("SAMPLE TAGGED SENTENCES FROM BROWN CORPUS:")
print("="*70)
for i, sentence in enumerate(train_data[:3], 1):
    print(f"\nSentence {i}:")
    print(sentence[:10])  # Show first 10 words

# 1. Train Unigram Tagger
print("\n" + "="*70)
print("TRAINING UNIGRAM TAGGER:")
print("="*70)
unigram_tagger = UnigramTagger(train_data)
unigram_accuracy = unigram_tagger.accuracy(test_data)
print(f"Unigram Tagger Accuracy: {unigram_accuracy:.4f} ({unigram_accuracy*100:.2f}%)")

# 2. Train Bigram Tagger with backoff
print("\n" + "="*70)
print("TRAINING BIGRAM TAGGER (with Unigram backoff):")
print("="*70)
bigram_tagger = BigramTagger(train_data, backoff=unigram_tagger)
bigram_accuracy = bigram_tagger.accuracy(test_data)
print(f"Bigram Tagger Accuracy: {bigram_accuracy:.4f} ({bigram_accuracy*100:.2f}%)")

# 3. Train Trigram Tagger with backoff chain
print("\n" + "="*70)
print("TRAINING TRIGRAM TAGGER (with Bigram & Unigram backoff):")
print("="*70)
trigram_tagger = TrigramTagger(train_data, backoff=bigram_tagger)
trigram_accuracy = trigram_tagger.accuracy(test_data)
print(f"Trigram Tagger Accuracy: {trigram_accuracy:.4f} ({trigram_accuracy*100:.2f}%)")

# Test on custom sentence
test_sentence = "The intelligent student studies machine learning"
test_tokens = word_tokenize(test_sentence)

print("\n" + "="*70)
print("TESTING CUSTOM TAGGERS ON NEW SENTENCE:")
print("="*70)
print(f"Sentence: {test_sentence}\n")

print("Unigram Tagger Result:")
print(unigram_tagger.tag(test_tokens))

print("\nBigram Tagger Result:")
print(bigram_tagger.tag(test_tokens))

print("\nTrigram Tagger Result:")
print(trigram_tagger.tag(test_tokens))
```

### ðŸ“Š Observations - Experiment 7:

1. **Unigram Tagger**:
   - Looks at each word independently
   - Typically 85-90% accuracy
   - Assigns most frequent tag for each word
   - Fast but ignores context
   
2. **Bigram Tagger**:
   - Considers previous word's tag
   - Better accuracy (90-93%)
   - Uses context from one previous word
   - Backoff to unigram for unknown sequences
   
3. **Trigram Tagger**:
   - Considers two previous tags
   - Highest accuracy (92-95%)
   - More context = better decisions
   - Chain of backoffs: Trigram â†’ Bigram â†’ Unigram
   
4. **Backoff Strategy**: Essential for handling unseen word sequences
5. **Training Data**: More training data = better accuracy
6. **Brown Corpus**: 1M+ words, diverse genres, excellent for training
7. **Trade-off**: Higher n-grams = more accuracy but more training data needed
8. **Practical Use**: Trigram with backoff is a good balance

## Experiment 8: POS Tagging for Text Processing Pipeline

Practical application: Using POS tagging in a complete text processing workflow.

```python
from collections import Counter

# Sample text for analysis
text = """
Python is an amazing programming language. It is widely used in data science, 
web development, and artificial intelligence. Many developers love Python because 
it is easy to learn and incredibly powerful. The syntax is clean and readable, 
making it perfect for beginners and experts alike.
"""

print("COMPLETE TEXT PROCESSING PIPELINE WITH POS TAGGING")
print("="*70)

# Step 1: Tokenization
tokens = word_tokenize(text)
print(f"\n1. TOKENIZATION:")
print(f"   Total tokens: {len(tokens)}")

# Step 2: POS Tagging
pos_tags = pos_tag(tokens)
print(f"\n2. POS TAGGING:")
print(f"   Tagged {len(pos_tags)} tokens")

# Step 3: Remove stopwords but keep important POS
stop_words = set(stopwords.words('english'))
important_pos = ['NN', 'NNS', 'NNP', 'NNPS', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ', 'JJ', 'JJR', 'JJS']

filtered_words = [
    (word, tag) for word, tag in pos_tags 
    if word.lower() not in stop_words and tag in important_pos
]

print(f"\n3. FILTERING (Remove stopwords, keep important POS):")
print(f"   Remaining tokens: {len(filtered_words)}")

# Step 4: Extract keywords (Nouns and Adjectives)
keywords = [word for word, tag in filtered_words if tag.startswith('NN') or tag.startswith('JJ')]
print(f"\n4. KEYWORD EXTRACTION (Nouns + Adjectives):")
print(f"   Keywords: {', '.join(keywords)}")

# Step 5: Frequency Analysis
word_freq = Counter(keywords)
print(f"\n5. FREQUENCY ANALYSIS:")
print(f"   Most common keywords:")
for word, count in word_freq.most_common(5):
    print(f"   - {word}: {count} times")

# Step 6: Extract action words (Verbs)
actions = [word for word, tag in filtered_words if tag.startswith('VB')]
print(f"\n6. ACTION EXTRACTION (Verbs):")
print(f"   Actions: {', '.join(actions)}")

# Step 7: POS Distribution
pos_distribution = Counter([tag for word, tag in pos_tags])
print(f"\n7. POS DISTRIBUTION:")
for pos, count in pos_distribution.most_common(10):
    print(f"   {pos}: {count} ({count/len(pos_tags)*100:.1f}%)")

# Step 8: Extract noun-adjective pairs
print(f"\n8. ADJECTIVE-NOUN PAIRS:")
for i in range(len(pos_tags)-1):
    word1, tag1 = pos_tags[i]
    word2, tag2 = pos_tags[i+1]
    if tag1.startswith('JJ') and tag2.startswith('NN'):
        print(f"   - {word1} {word2}")
```

### ðŸ“Š Observations - Experiment 8:

1. **Pipeline Integration**: POS tagging fits naturally into NLP workflows
2. **Smart Filtering**: Can remove stopwords while preserving important content words based on POS
3. **Keyword Extraction**: Nouns and adjectives make excellent keywords
4. **Content Analysis**: Verb extraction reveals actions and processes discussed
5. **Frequency Patterns**: POS distribution reveals text characteristics
6. **Phrase Detection**: Sequential POS tags help identify meaningful phrases (adj-noun pairs)
7. **Efficiency**: POS-based filtering is more intelligent than simple stopword removal
8. **Applications**:
   - Document summarization
   - Information extraction
   - Search engine optimization
   - Content categorization
9. **Advantage**: POS tagging enables semantic understanding beyond word frequency

## Experiment 10: Named Entity Recognition with POS Tagging

Combining POS tagging with chunking for Named Entity Recognition.

```python
from nltk import ne_chunk, RegexpParser

# Sample text with named entities
text = """
Elon Musk is the CEO of Tesla and SpaceX. He was born in South Africa and 
later moved to the United States. In 2022, he acquired Twitter for 44 billion 
dollars. Apple Inc. and Microsoft Corporation are also major technology companies.
"""

# Tokenize and POS tag
tokens = word_tokenize(text)
pos_tags = pos_tag(tokens)

print("NAMED ENTITY RECOGNITION WITH POS TAGGING")
print("="*70)
print(f"Text: {text.strip()}\n")

# Method 1: Using NLTK's built-in NER
print("\n1. BUILT-IN NAMED ENTITY RECOGNITION:")
print("-"*70)
named_entities = ne_chunk(pos_tags)
print(named_entities)

# Extract named entities
print("\n2. EXTRACTED NAMED ENTITIES:")
print("-"*70)
for chunk in named_entities:
    if hasattr(chunk, 'label'):
        entity_name = ' '.join(c[0] for c in chunk)
        entity_type = chunk.label()
        print(f"   {entity_name:30} --> {entity_type}")

# Method 2: Custom chunking using POS patterns
print("\n3. CUSTOM NOUN PHRASE CHUNKING:")
print("-"*70)

# Define grammar for noun phrases
grammar = r"""
    NP: {<DT>?<JJ>*<NN.*>+}          # Noun phrases
    PERSON: {<NNP>+}                   # Person names (consecutive proper nouns)
    ORGANIZATION: {<NNP>+<NN.*>}       # Organizations
"""

# Create parser
chunk_parser = RegexpParser(grammar)
chunks = chunk_parser.parse(pos_tags)

print("Noun Phrases and Named Entities:")
for subtree in chunks.subtrees():
    if subtree.label() != 'S':  # Skip sentence root
        phrase = ' '.join(word for word, tag in subtree.leaves())
        print(f"   {phrase:35} --> {subtree.label()}")

# Method 3: Extract proper nouns (potential named entities)
print("\n4. ALL PROPER NOUNS (NNP/NNPS):")
print("-"*70)
proper_nouns = [word for word, tag in pos_tags if tag in ['NNP', 'NNPS']]
print(f"   {', '.join(proper_nouns)}")

# Group consecutive proper nouns
print("\n5. CONSECUTIVE PROPER NOUN SEQUENCES:")
print("-"*70)
i = 0
while i < len(pos_tags):
    if pos_tags[i][1] == 'NNP':
        entity = [pos_tags[i][0]]
        j = i + 1
        while j < len(pos_tags) and pos_tags[j][1] == 'NNP':
            entity.append(pos_tags[j][0])
            j += 1
        print(f"   {' '.join(entity)}")
        i = j
    else:
        i += 1
```

### ðŸ“Š Observations - Experiment 10:

1. **POS Foundation**: NER heavily relies on POS tagging (especially proper nouns)
2. **Built-in NER**: NLTK identifies PERSON, ORGANIZATION, GPE (geo-political entities)
3. **Consecutive NNP**: Grouping consecutive proper nouns reveals multi-word entities
4. **Chunking Grammar**: Regex patterns over POS tags enable custom entity extraction
5. **Patterns**:
   - Person names: Multiple consecutive NNP tags
   - Organizations: NNP + NN patterns (e.g., "Apple Inc.")
   - Locations: Typically NNP tags
6. **Limitations**: Simple POS-based NER can miss context-dependent entities
7. **Applications**:
   - Information extraction from documents
   - Building knowledge graphs
   - Content categorization
   - Question answering systems
8. **Enhancement**: Modern NER uses deep learning, but POS provides baseline features

## Experiment 11: Handling Misspellings and Unknown Words

See how POS taggers handle spelling errors and unknown words.

```python
# Test sentences with various challenges
test_cases = [
    ("The cat is sleeping peacefully.", "Correct spelling"),
    ("The kat is sleping peacfully.", "Multiple misspellings"),
    ("I xyzabc the beautiful garden.", "Nonsense word"),
    ("Gpt4 and ChatGPT are revolutionizing AI.", "New tech terms"),
    ("She eated the pizza quickly.", "Incorrect verb form"),
    ("The flibbertigibbet was very confuzzled.", "Rare/made-up words")
]

print("HANDLING MISSPELLINGS AND UNKNOWN WORDS")
print("="*70)

for sentence, description in test_cases:
    print(f"\n{description.upper()}:")
    print(f"Sentence: {sentence}")
    print("-"*70)
    
    tokens = word_tokenize(sentence)
    tags = pos_tag(tokens)
    
    print("POS Tags:")
    for word, tag in tags:
        print(f"   {word:20} --> {tag}")
    
    # Check if any words are tagged as NN (common fallback for unknown words)
    unknown_indicators = [word for word, tag in tags if tag == 'NN' and word.lower() not in ['cat', 'garden', 'pizza', 'words']]
    if unknown_indicators:
        print(f"Possible unknown words: {', '.join(unknown_indicators)}")
```

### ðŸ“Š Observations - Experiment 11:

1. **Robust Handling**: POS taggers attempt to tag unknown words based on context
2. **Misspellings**: Often tagged based on position and surrounding words
3. **Common Fallback**: Unknown words frequently tagged as NN (noun) by default
4. **Context Clues**: Tagger uses surrounding words to make educated guesses
5. **New Terms**: Technical terms (like "GPT4") may be tagged as NNP or NN
6. **Morphology**: Word endings help (e.g., "-ly" suggests adverb)
7. **Limitations**: 
   - Cannot fix spelling automatically
   - May mis-tag severely misspelled words
   - New slang or jargon can confuse tagger
8. **Real-world Impact**: Important to clean text before POS tagging for best accuracy
9. **Preprocessing**: Spell-checking should precede POS tagging in production pipelines

## Experiment 12: Real-World Application - Sentiment Analysis Enhancement

Use POS tagging to improve sentiment analysis by focusing on opinion-bearing words.

```python
# Sample reviews for sentiment analysis
reviews = [
    "This product is absolutely amazing! The quality is excellent and delivery was fast.",
    "Terrible experience. The item was broken and customer service was unhelpful.",
    "It's okay. Nothing special but it works as expected.",
    "I love it! Best purchase ever. Highly recommend to everyone.",
    "Disappointed. Poor quality and overpriced. Would not buy again."
]

# Define sentiment words (simplified)
positive_words = ['amazing', 'excellent', 'fast', 'love', 'best', 'recommend', 'great', 'wonderful', 'fantastic']
negative_words = ['terrible', 'broken', 'unhelpful', 'disappointed', 'poor', 'overpriced', 'bad', 'awful', 'horrible']

print("SENTIMENT ANALYSIS WITH POS TAGGING")
print("="*70)

for i, review in enumerate(reviews, 1):
    print(f"\nREVIEW {i}: {review}")
    print("-"*70)
    
    tokens = word_tokenize(review.lower())
    pos_tags = pos_tag(tokens)
    
    # Extract adjectives and adverbs (opinion-bearing words)
    opinion_words = [
        (word, tag) for word, tag in pos_tags 
        if tag.startswith('JJ') or tag.startswith('RB')
    ]
    
    print("Opinion-bearing words (Adjectives & Adverbs):")
    for word, tag in opinion_words:
        print(f"   {word:20} ({tag})")
    
    # Simple sentiment scoring
    positive_count = sum(1 for word, _ in opinion_words if word in positive_words)
    negative_count = sum(1 for word, _ in opinion_words if word in negative_words)
    
    # Calculate sentiment
    if positive_count > negative_count:
        sentiment = "POSITIVE âœ“"
    elif negative_count > positive_count:
        sentiment = "NEGATIVE âœ—"
    else:
        sentiment = "NEUTRAL â—"
    
    print(f"\nSentiment Score: +{positive_count} positive, -{negative_count} negative")
    print(f"Overall Sentiment: {sentiment}")
    print("="*70)

# Advanced analysis: Negation handling
print("\n\nADVANCED: NEGATION HANDLING WITH POS")
print("="*70)

negation_sentence = "The product is not good but the service is not bad either"
tokens = word_tokenize(negation_sentence.lower())
pos_tags = pos_tag(tokens)

print(f"Sentence: {negation_sentence}")
print("\nAnalyzing with negation detection:")

negation_words = ['not', 'no', "n't", 'never', 'neither']
negated = False

for i, (word, tag) in enumerate(pos_tags):
    if word in negation_words:
        negated = True
        print(f"   NEGATION detected: '{word}'")
    elif tag.startswith('JJ') or tag.startswith('RB'):
        status = "NEGATED" if negated else "DIRECT"
        print(f"   Opinion word: '{word}' ({tag}) - {status}")
        negated = False  # Reset after opinion word
```

### ðŸ“Š Observations - Experiment 12:

1. **Opinion Focus**: Adjectives (JJ) and adverbs (RB) are primary sentiment carriers
2. **POS Filtering**: Reduces noise by focusing only on opinion-bearing word classes
3. **Efficiency**: POS-based filtering is faster than analyzing all words
4. **Accuracy Boost**: Ignoring nouns/verbs often irrelevant to sentiment
5. **Negation Handling**: POS helps identify opinion words that follow negations
6. **Intensifiers**: Adverbs like "absolutely", "extremely" modify sentiment strength
7. **Context Awareness**: "not good" vs "good" shows importance of word sequence
8. **Applications**:
   - Product review analysis
   - Social media monitoring
   - Customer feedback processing
   - Brand reputation management
9. **Limitation**: Simple word-based sentiment needs enhancement with deep learning
10. **Best Practice**: Combine POS tagging with sentiment lexicons for better results
